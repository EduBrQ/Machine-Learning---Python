```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
salario_edu = np.array([100, 200, 300, 500, 400, 150])
salario_cass = np.array([50, 100, 60, 350, 100, 600])
salario_fulano = np.array([80, 200, 90, 230, 300, 500])
salario = (salario_edu + salario_cass)
salario
```

```python
plt.plot(salario)
```

```python
# colocando um label 
plt.plot(salario_edu, c='Black', ls='--', marker='s', ms=8, label='Salario Eduardo') 
plt.plot(salario_cass, c='Red', ls='--', marker='^', ms=8, label='Salario Cassandra') 
plt.plot(salario_fulano, c='Blue', ls='--', marker='o', ms=8, label='Salario Fulano') 

# colocando uma legenda
plt.legend(loc = 'upper left')
plt.show() # nao mostra a linha -> [<matplotlib.lines.Line2D at 0x9562190>]
```
